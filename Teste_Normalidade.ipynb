# **Teste de Normalidade**

```python
# Importando pacotes
from scipy import stats
from scipy.stats import norm, skewnorm
import matplotlib.pyplot as plt
```

```python
# Criação de uma variavel com dados de uma distribuiçã normal com função rvs (100 elementos)
dados = norm.rvs(size = 1000)
```

```python
#histograma
plt.hist(dados, bins = 20)
plt.title('Dados')
```

```python
# Geração de grafico para verificar se a distribuição é normal
fig, ax = plt.subplots()
stats.probplot(dados, fit=True, plot=ax)
plt.show()
```

```python
#Execução do Teste de shapiro
#Segundo o argumento é o valor de p, não há como rejeitar a hipótese nula
stats.shapiro(dados)
```

### Dados não normais

```python
dados2 = skewnorm.rvs(4, size=1000)
```

```python
#Histograma
plt.hist(dados2, bins = 20)
plt.title('Dados')
```

```python
#Geração de grafico para verificar se a distribuição é  normal
fig, ax = plt.subplots()
stats.probplot(dados2, fit =True, plot= ax)
plt.show()
```

```python
#Teste de shapiro
stats.shapiro(dados2)
```
